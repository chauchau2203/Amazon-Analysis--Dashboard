### III. Data Cleaning

#### 3.1 Amazon Sales Report

Dropping unecessary columns

```python
amz_columns_to_drop = ['index', 'Unnamed: 22', 'fulfilled-by', 'ship-country', 'currency', 'Sales Channel']
existing_amz_columns_to_drop = [col for col in amz_columns_to_drop if col in amazon_sales.columns]
amazon_sales.drop(columns = existing_amz_columns_to_drop, inplace = True)
```

Dropping duplicated values

```python
# Viewing duplicated values
amazon_sales[amazon_sales.duplicated(['Order ID','ASIN'], keep = False)]
```

```python
# Dropping duplicates
amazon_sales.drop_duplicates(['Order ID','ASIN'], inplace = True, ignore_index = True)
```

Handling NaN values

```python
# Replacing missing values in 'Courier Status' with 'Unknown'
amazon_sales.fillna({'Courier Status':'Unknown'}, inplace = True)

# Replaceing missing values in 'promotion-ids' with 'No Promotion'
amazon_sales.fillna({'promotion-ids':'No Promotion'}, inplace = True)

# Replacing missing values in 'Amount' with 0
amazon_sales.fillna({'Amount': 0}, inplace=True)

# Replacing missing values in 'ship-city' with 'Unknown'
amazon_sales.fillna({'ship-city':'Unknown'}, inplace = True)

# Replacing missing values in 'ship-state' with 'Unknown'
amazon_sales.fillna({'ship-state':'Unknown'}, inplace = True)

# Replacing missing values in 'ship-postal-code' with 'Unknown'
amazon_sales.fillna({'ship-postal-code':'Unknown'}, inplace = True)

# Replacing missing values in 'Size' with 'FREE'
amazon_sales.fillna({'Size':'FREE'}, inplace = True)
```

Renaming columns

```python
amz_col_names = {'Order ID':'order_ID', 'Date':'date', 'Status':'ship_status','Fulfilment':'fullfilment',
          'ship-service-level':'service_level', 'Style':'style', 'SKU':'sku', 'Category':'product_category', 
          'Size':'size', 'ASIN':'asin', 'Courier Status':'courier_ship_status', 'Qty':'order_quantity', 
          'Amount':'order_amount_($)', 'ship-city':'city', 'ship-state':'state', 'ship-postal-code':'zip', 
          'promotion-ids':'promotion','B2B':'customer_type'}
amazon_sales.rename(columns = amz_col_names, inplace = True)
```

Dropping invalid values in the city column

```python
# Define a list of invalid or error values commonly found in city columns
invalid_cities = [',Hyderabad', '53Miles', '7Barasat', '(Chikmagalur Disterict).     (N.R Pur Thaluku)', '6Th Mile Tadong', '.Gannavaram', '.Azamgarh','Ã Â¤Â²Ã Â¤â€“Ã Â¤Â¨Ã Â¤Å ', 
                  '147/19 B Keshab Chandra Sen Street Kolkata Nine Ma', '..Katra', '116  B. P. M. B Sarani, Uttarpara Kotrung', '(Via Cuncolim)Quepem,South Goa', '1', ',Raibarely Road Faizabad (Ayodhya)']

# Drop rows with invalid city values
amazon_sales = amazon_sales[~amazon_sales['city'].isin(invalid_cities)]
```

Converting INR currency to USD currency in the Amount column (exchange rate of 1 INR = 0.012 USD)

```python
exchange_rate = 0.0120988
amazon_sales['order_amount_($)'] = (amazon_sales['order_amount_($)'] * exchange_rate).round(2)
```

Converting values in the B2B column

```python
amazon_sales['customer_type'].replace(to_replace = [True, False], value = ['Business', 'Customer'], inplace = True)
```

Creating Datetime and Adding a Month column

```python
# Making sure the 'date' column is in datetime format
amazon_sales['date'] = pd.to_datetime(amazon_sales['date'])

# Dropping March dates from the dataset as there is only one day for March
amazon_sales = amazon_sales[(amazon_sales['date'].dt.month != 3)]

# Extracting month from the date
amazon_sales['date'] = pd.to_datetime(amazon_sales['date'])
amazon_sales['month'] = amazon_sales['date'].dt.month

# Change month number to month name
month_name = {4: 'April', 5: 'May', 6: 'June'}
amazon_sales['month'] = amazon_sales['date'].dt.month.map(month_name)

# Convert the month column to a categorical data type with the desired order
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
amazon_sales['month'] = pd.Categorical(amazon_sales['month'], categories = month_order, ordered = True)
```

Column Value Ordering

```python
# Create an ordered categorical variable for the 'size' column
size_order = ['FREE', 'XS', 'S', 'M', 'L', 'XL', 'XXL', 'XXXL', '3XL', '4XL', '5XL', '6XL',]
amazon_sales['size'] = pd.Categorical(amazon_sales['size'], categories = size_order, ordered = True)
```

Normalising text columns (converting to Title Case) and Removing white spaces

```python
amz_text_columns = ['ship_status', 'fullfilment', 'product_category', 'city', 'state', 'month']  

for column in amz_text_columns:
    amazon_sales[column] = amazon_sales[column].str.strip().str.title()  

amazon_sales['size'] = amazon_sales['size'].str.strip().str.upper()  

# Handle special cases like New Delhi
amazon_sales.loc[amazon_sales['city'] == 'New Delhi', 'state'] = 'Delhi'
```

Viewing the data after cleaned

```python
amazon_sales.head()
```

Exporting the cleaned dataset

```python
# Exporting the cleaned dataset to a CSV file
amazon_sales.to_csv("E:/PYTHON/Amazon India Ecommerce Sales Dataset/Cleaned dataset/amazon_sales.csv", index=False)
```

#### 3.4 International Sales

In terms of verticle structure, the file contains 3 parts, separated at index 18634 and 19674. The first and third part contain data of the same orders. The second part contains a list of products and their stock (displayed at the end of this part).\
\
Separate the file into three dataframes

```python
# Create the first dataframe (index 0 to 18634)
international_sales_1 = international_sales[international_sales['index'].between(0, 18634)].reset_index(drop = True)

# Create the second dataframe (index 18660 to 19674). Rows from index 18635 to 18659 was skipped due to missing values in stock
international_sales_2 = international_sales[international_sales['index'].between(18660, 19674)].reset_index(drop = True)

# Create the third dataframe (remaining part)
international_sales_3 = international_sales[international_sales['index'] > 19675].reset_index(drop = True)
```

Dropping unecessary columns

```python
# Viewing the NaN values in the second dataframe
international_sales_2.isnull().sum()
```

```python
# Dropping columns in 'international_sales_2' as they only have NaN values
international_sales_2.drop(columns = ['index', 'CUSTOMER', 'Style', 'SKU', 'Size', 'PCS', 'RATE', 'GROSS AMT'], 
                           inplace = True)

# Dropping columns in 'international_sales_1' and 'international_sales_3'
international_sales_1.drop(columns = ['index'], inplace = True)
international_sales_3.drop(columns = ['index'], inplace = True)
```

Renaming the columns

```python
# Renaming columns in international_sales_1
international_sales_1.rename(columns={
    'DATE': 'date',
    'Months': 'month',
    'CUSTOMER': 'customer',
    'Style': 'style',
    'SKU': 'sku',
    'Size': 'size',
    'PCS': 'quantity',
    'RATE': 'rate($)',
    'GROSS AMT': 'gross_amount($)'
}, inplace=True)

# Renaming columns in international_sales_2
international_sales_2.rename(columns={
    'DATE': 'style',
    'Months': 'stock'
}, inplace=True)

#Renaming columns in international_sales_3
international_sales_3.rename(columns={
    'DATE': 'customer',
    'Months': 'date',
    'CUSTOMER': 'month',
    'Style': 'style',
    'SKU': 'sku',
    'Size': 'quantity',
    'PCS': 'rate($)',
    'RATE': 'gross_amount($)',
    'GROSS AMT': 'stock'
}, inplace=True)
```

Normalising and cleaning the text columns

```python
# Converting the 'customer' column in 'international_sales_1' to Title Case and strip white spaces
international_sales_1['customer'] = international_sales_1['customer'].str.strip().str.title()

# Converting the 'customer' column in 'international_sales_3' to Title Case and strip white spaces
international_sales_3['customer'] = international_sales_3['customer'].str.strip().str.title()
```

Column Value Ordering

```python
international_sales_1['size'] = pd.Categorical(international_sales_1['size'], 
                                               categories = size_order, ordered = True)
```

Converting columns to numerical values

```python
# 'international_sales_1' numerical columns
int_1_num_cols = ['quantity', 'rate($)', 'gross_amount($)']
for column in int_1_num_cols:
    international_sales_1[column] = international_sales_1[column].astype(float)    

# 'international_sales_3' numerical columns
int_3_num_cols = ['quantity', 'rate($)', 'gross_amount($)', 'stock']
for column in int_3_num_cols:
    international_sales_3[column] = international_sales_3[column].astype(float)
```

Converting INR currency to USD currency

```python
# 'international_sales_1' currency columns
international_sales_1['rate($)'] = (international_sales_1['rate($)'] * exchange_rate).round(2)
international_sales_1['gross_amount($)'] = (international_sales_1['gross_amount($)'] * exchange_rate).round(2)

# 'international_sales_3' currency columns
international_sales_3['rate($)'] = (international_sales_3['rate($)'] * exchange_rate).round(2)
international_sales_3['gross_amount($)'] = (international_sales_3['gross_amount($)'] * exchange_rate).round(2)
```

Filling in missing values

```python
# Viewing null values in 'international_sales_1'
international_sales_1.isnull().sum()
```

```python
# Viewing null values in 'international_sales_3'
international_sales_3.isnull().sum()
```

```python
# Replacing missing values in 'international_sales_1' with NA
international_sales_1.fillna({'sku':'NA'}, inplace = True)

# Replacing missing values in 'international_sales_3' with NA
international_sales_3.fillna({'sku':'NA'}, inplace = True)
```

Formatting the date, month and year column

```python
# Making sure the 'date' column is in datetime format
international_sales_1['date'] = pd.to_datetime(international_sales_1['date'])
international_sales_3['date'] = pd.to_datetime(international_sales_3['date'])

# Change 'month' column values into only months
international_sales_1['month'] = international_sales_1['date'].dt.strftime('%B').str.title()
international_sales_3['month'] = international_sales_3['date'].dt.strftime('%B').str.title()

# Extracting year from the 'date' column
international_sales_1['year'] = international_sales_1['date'].dt.year
international_sales_3['year'] = international_sales_3['date'].dt.year
```

```python
# Convert the month column to a categorical data type with the desired order
international_sales_1['month'] = pd.Categorical(international_sales_1['month'], 
                                                categories = month_order, ordered = True)
international_sales_3['month'] = pd.Categorical(international_sales_3['month'], 
                                                categories = month_order, ordered = True)
```
